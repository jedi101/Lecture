# Merkmalskalierung

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from numpy.random import default_rng
from sklearn import datasets, linear_model, preprocessing
```

Generiere zufälligen Datensatz mit zwei Merkmalen $x1$ und $x2$.

```python
rng = default_rng(seed=42)
ds_x1 = 200 * rng.random(100) + 100
ds_x2 = 2 * rng.random(100)

x1 = ds_x1
x2 = ds_x2

plt.scatter(x1, x2)
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
```

```python
min_x = np.min(x1)
max_x = np.max(x1)
min_y = np.min(x2)
max_y = np.max(x2)
print("Variable x1 bewegt sich im Wertebereich [" + str(min_x) + ", " + str(max_x) + "]")
print("Variable x2 bewegt sich im Wertebereich [" + str(min_y) + ", " + str(max_y) + "]")
```

## Min-Max Skalierung

Für die Min-Max Skalierung der Daten wird die folgende Formel verwendet:

$$x' = \frac{x - x_{min}}{x_{max} - x_{min}}$$

```python
x1 = ds_x1
x2 = ds_x2

plt.scatter(x1, x2, color='b')
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()

# Hinweis: Sie können np.min() und np.max() benutzen
# -- start of your code



# -- end of your code

plt.scatter(x1, x2, color='g')
plt.xlabel("x1 Min-Max Skaliert")
plt.ylabel("x2 Min-Max Skaliert")
plt.show()
```
